```python
import folium
import json
import os
import pandas as pd
from IPython.display import HTML
```

```python
df = pd.read_csv('./data/stations-velib.csv', delimiter=";")
df.head(2)
```

```python
df[['longitude','latitude']] = df['position'].str.split(',',expand=True).astype(float)
df['zipcode'] = df['address'].str.extract('(\d{5})',expand=True)
df = df[df['zipcode'] == "75015"]
df.head()
```

```python
mymap = folium.Map(location=[48.839635,2.300654],zoom_start=14)
```

```python
for index, station in df.iterrows():
    folium.Marker([station['longitude'], station['latitude']],popup=station['name']).add_to(mymap)
mymap
```

```python
regions_json = json.load(open('./data/regions.json'))
regions_map = folium.Map(location=[48,2],zoom_start=5)
folium.GeoJson(regions_json,name='hello').add_to(regions_map)
```

```python
df = pd.read_csv('./data/merged_named_regions.csv')
df = df[df['Année']==2011]
df['insee_region'] = df['insee_region'].apply(str)
```

```python
sun = folium.Map(location=[48,2], zoom_start=5)
```

```python
sun.choropleth(
    geo_data='./data/regions.json',
    name='choropleth',
    data=df,
    columns=['insee_region', 'Production solaire renouvelable (GWh)'],
    key_on='feature.properties.code',
    fill_color='YlOrRd',
    fill_opacity=0.7,
    line_opacity=0.5,
    legend_name='Solar production'
)
folium.LayerControl().add_to(sun)
```

```python
sun.save("./data/sun.html")
```
