# List S&P 500 companies


```python
import bs4 as bs
import pickle
import requests
```

```python
response = requests.get('https://en.wikipedia.org/wiki/List_of_S%26P_500_companies')
soup = bs.BeautifulSoup(response.text,"lxml")
table = soup.find('table',{'class':'wikitable sortable'})
companies = []
for row in table.findAll('tr')[1:]:
    company = row.findAll('td')
    if company[3].text == "Energy":
        companies.append(company[0].text)
print(companies)
```

# Get stock prices for all companies

```python
start = dt.datetime(2017,1,1)
end = dt.datetime(2017,11,1)
for company in companies:
    try:
        df = web.DataReader(company,'google',start, end)
        df.to_csv('./data/stock/{}.csv'.format(company))
    except:
        
        print("No data for:",company)
```

# Combine all data

```python
import glob
import os
import re
main_df = pd.DataFrame()
for file in glob.glob('./data/stock/*.csv'):
    df = pd.read_csv(file)
    df.set_index('Date',inplace=True)
    df.rename(columns={'Close':re.findall('[A-Z]+', file)[0]},inplace=True)
    df.drop(['Open','High','Low','Volume'],1, inplace=True)
    if main_df.empty:
        main_df = df
    else:
        main_df = main_df.join(df,how='outer')
```
