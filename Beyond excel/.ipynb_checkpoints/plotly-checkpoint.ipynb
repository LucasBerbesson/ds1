```python
import requests
import pandas as pd
import plotly.graph_objs as go
from plotly.offline import plot
```

```python
response = requests.get(
        "https://opendata.reseaux-energies.fr/api/records/1.0/search/?dataset=equilibre-national-mensuel-prod-conso-brute&rows=100&facet=mois").json()
df = pd.DataFrame([x['fields'] for x in response["records"]])
df["mois"] = pd.to_datetime(df["mois"],format="%Y-%m")
df.set_index("mois",inplace=True)
df.sort_index(inplace=True)

data = [ go.Scatter(x=df.index,y=df["production_totale"],fillcolor="#a6bddb",name="Max capacity price"),
        go.Scatter(x=df.index,y=df["consommation_brute"],fillcolor = "#a6bddb",name = "Max capacity price")]

plot(data)
```

```python
response = requests.get(
        "https://opendata.reseaux-energies.fr/api/records/1.0/search/?dataset=extremas-quotidiens-flux-commerciaux&rows=100&facet=date").json()
df = pd.DataFrame([x['fields'] for x in response["records"]])
df["date"] = pd.to_datetime(df["date"],format="%Y-%m-%d")
df.set_index("date",inplace=True)
df.sort_index(inplace=True)

data = [ go.Scatter(x=df.index,y=df["solde_max"],fillcolor="#a6bddb",name="Solde max"),
        go.Scatter(x=df.index,y=df["solde_min"],fillcolor = "#a6bddb",name = "Solde min")]

plot(data)
```
