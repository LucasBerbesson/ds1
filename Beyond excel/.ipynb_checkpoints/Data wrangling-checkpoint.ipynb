# Fusionner des données
Fusionner consiste à rassembler les lignes de deux DataFrames en utilisant une clé commune. 

```python
import pandas as pd
```

```python
productions = pd.read_csv('./data/merged.csv')
```

```python
print("Dataset de", productions.shape[0], "lignes")
productions.head()
```

```python
insee_codes = pd.read_csv('./data/insee_regions.csv', names=['Region','region_code'])
```

```python
print("Dataset de", insee_codes.shape[0], "lignes")
insee_codes.head()
```

```python
df = pd.merge(productions, insee_codes,how='left',left_on="insee_region",right_on="region_code")
print("Dataset de", insee_codes.shape[0], "lignes")
df.head()
```
