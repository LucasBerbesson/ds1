```python
import json
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('PriceData.csv',index_col=0)
df.head()
```

```python
df.index = pd.to_datetime(df.index, format='%d/%m/%y')
```

```python
df.iloc[:,0:3].dropna().plot(figsize=(15,8))
```

```python
import plotly.offline as py
py.init_notebook_mode(connected=True)
import plotly.graph_objs as go

data = [go.Scatter( x=df.index, y=df[column].dropna(), name=column) for column in df.columns[:6]]    
py.iplot(data)
```

```python
df['ZEE_DA_EUR_MA_20'] = df['ZEE_DA_EUR'].rolling(window=20, min_periods=15).mean()
df['ZEE_DA_EUR_STD_20_sup'] = df['ZEE_DA_EUR_MA_20'] + 2*df['ZEE_DA_EUR'].rolling(window=20, min_periods=15).std()
df['ZEE_DA_EUR_STD_20_inf'] = df['ZEE_DA_EUR_MA_20'] - 2*df['ZEE_DA_EUR'].rolling(window=20, min_periods=15).std()
```

```python
data = [go.Scatter( x=df.index, y=df['ZEE_DA_EUR'], name='ZEE_DA_EUR'),
        go.Scatter( x=df.index, y=df['ZEE_DA_EUR_MA_20'], name='ZEE_DA_EUR_MA_20'),
        go.Scatter( x=df.index, y=df['ZEE_DA_EUR_STD_20_sup'], name='ZEE_DA_EUR_STD_20_sup',marker={'color':'pink'}),
        go.Scatter( x=df.index, y=df['ZEE_DA_EUR_STD_20_inf'], name='ZEE_DA_EUR_STD_20_inf',marker={'color':'pink'},fill='tonexty'),
       ]    
py.iplot(data)
```
